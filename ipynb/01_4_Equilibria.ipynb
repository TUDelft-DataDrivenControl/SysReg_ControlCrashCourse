# Equilibrium points

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;*Forever and unchanging*

Often, systems will have equilibria, meaning there are states where system will remain over time. These state-input point pairs are denoted as $(x_e, u_e)$ or $(\bar x, \bar u)$. I'll use the latter since this is more traditional in control and we have to appease our control elders. So equilibria remain unchanging, so the derivative of the state is zero. In mathematicians' language
$$ \dot{\bar x} = f(\bar x, \bar u) = 0. $$
Finding the equilibria states as a function of the equilibria inputs is as simple as solving this equation.
